```python
from knn import MyKNN
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

class1_vecs = np.random.normal(loc=[1, 1], scale=[0.5, 1], size=(100, 2))
class2_vecs = np.random.normal(loc=[-1, -1], scale=[1, 0.5], size=(100, 2))
class3_vecs = np.random.normal(loc=[-1, 2], scale=[1, 1.5], size=(100, 2))
all_vecs = np.vstack((class1_vecs, class2_vecs, class3_vecs))

class1_labels = np.full((100,), 'Class 1')
class2_labels = np.full((100,), 'Class 2')
class3_labels = np.full((100,), 'Class 3')
all_labels = np.hstack((class1_labels, class2_labels, class3_labels))

indices = np.random.permutation(all_vecs.shape[0])
data = all_vecs[indices]
labels = all_labels[indices]
```

```python
plt.figure(figsize=(10, 8))

for label, color in zip(['Class 1', 'Class 2', 'Class 3'], ['red', 'blue', 'green']):
    mask = labels == label
    plt.scatter(data[mask][:, 0], data[mask][:, 1], label=label, c=color, alpha=0.6)

plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
knn = MyKNN(data=data, labels=labels, k=5)
input = np.array([0, -0.5])
```

```python
knn.knn_search(input)
```

```python
knn.classify(input)
```

```python
plt.figure(figsize=(10, 8))

for label, color in zip(['Class 1', 'Class 2', 'Class 3'], ['red', 'blue', 'green']):
    mask = labels == label
    plt.scatter(data[mask][:, 0], data[mask][:, 1], label=label, c=color, alpha=0.6)


plt.scatter(input[0], input[1], label='Test', c='black', alpha=0.5, s=150)

plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.grid(True)
plt.show()
```
